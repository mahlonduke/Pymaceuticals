## Observations and Insights

## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import datetime

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
```

```python
# Sample of mouse_metadata
mouse_metadata.head()
```

```python
# Sample of study_results
study_results.head()
```

```python
# Combine the data into a single dataset
mergedSource = pd.merge(mouse_metadata, study_results, on="Mouse ID")
mergedSource.head()
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume 
# for each regimen
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug 
mean = mergedSource.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
mean = pd.DataFrame(mean)

# Flatten the column headings
mean = mean.reset_index()

# Rename the column to differentiate from the other calculations
mean = mean.rename(columns={"Tumor Volume (mm3)": "Tumor Volume - Mean"})

# Preview DataFrame
mean
```

```python
# Store the Median Tumor Volume Data Grouped by Drug 
median = mergedSource.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]

# Convert to DataFrame
median = pd.DataFrame(median)

# Flatten the column headings
median = median.reset_index()

# Rename the column to differentiate from the other calculations
median = median.rename(columns={"Tumor Volume (mm3)": "Tumor Volume - Median"})

# Preview DataFrame
median
```

```python
# Store the Mode Tumor Volume Data Grouped by Drug 
#mode = mergedSource.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].agg(['Tumor Volume (mm3)'].mode)
#print(mode)


mode = mergedSource.groupby(['Drug Regimen']).agg(lambda x:x.value_counts().index[0])["Tumor Volume (mm3)"]

# Convert to DataFrame
mode = pd.DataFrame(mode)

# Flatten the column headings
mode = mode.reset_index()

# Rename the column to differentiate from the other calculations
mode = mode.rename(columns={"Tumor Volume (mm3)": "Tumor Volume - Mode"})

# Preview DataFrame
mode
```

```python
# Store the variance Tumor Volume Data Grouped by Drug 
variance = mergedSource.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]

# Convert to DataFrame
variance = pd.DataFrame(variance)

# Flatten the column headings
variance = variance.reset_index()

# Rename the column to differentiate from the other calculations
variance = variance.rename(columns={"Tumor Volume (mm3)": "Tumor Volume - Variance"})

# Preview DataFrame
variance
```

```python
# Store the standard deviation Tumor Volume Data Grouped by Drug 
standardDeviation = mergedSource.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]

# Convert to DataFrame
standardDeviation = pd.DataFrame(standardDeviation)

# Flatten the column headings
standardDeviation = standardDeviation.reset_index()

# Rename the column to differentiate from the other calculations
standardDeviation = standardDeviation.rename(columns={"Tumor Volume (mm3)": "Tumor Volume - Standard Deviation"})

# Preview DataFrame
standardDeviation
```

```python
# Store the standard error Tumor Volume Data Grouped by Drug 
standardError = mergedSource.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
standardError = pd.DataFrame(standardError)

# Flatten the column headings
standardError = standardError.reset_index()

# Rename the column to differentiate from the other calculations
standardError = standardError.rename(columns={"Tumor Volume (mm3)": "Tumor Volume - Standard Error"})

# Preview DataFrame
standardError
```

```python
# Merge everything

# Create the analysis dataframe
analysis_df = {}
pd.DataFrame(analysis_df)

# Merge data into the analysis dataframe
analysis_df = pd.merge(mean, median, on="Drug Regimen")
analysis_df = pd.merge(analysis_df, mode, on="Drug Regimen")
analysis_df = pd.merge(analysis_df, variance, on="Drug Regimen")
analysis_df = pd.merge(analysis_df, standardDeviation, on="Drug Regimen")
analysis_df = pd.merge(analysis_df, standardError, on="Drug Regimen")
analysis_df
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. 
    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
mergedSource.head()
```

```python
#Filter the source for just the final timepoints
final = mergedSource['Timepoint']==45
finalTimepoints = mergedSource[final]
finalTimepoints
```

```python
# Sort the filtered dataframe by tumor volume to find the four drugs which resulted in the smallest tumors.  
sorted = finalTimepoints.sort_values(by=['Tumor Volume (mm3)'])
leadingDrugs = ["Ramicane", "Capomulin", "Propriva", "Placebo"]
sorted.head(50)

```

```python
# Rename the Drug Regimen column to make it easier to call as a function
renamedTimepoints = finalTimepoints.rename(columns={"Drug Regimen": "Drug"})
renamedTimepoints
```

```python
# Filter the dataframe containing final timepoints by the leading drugs
filteredLeadingDrugs = renamedTimepoints.Drug.isin(leadingDrugs)

filteredLeadingDrugs = renamedTimepoints[filteredLeadingDrugs]
filteredLeadingDrugs.head()
```

```python
# Calculate the final tumor volumes
finalVolume = filteredLeadingDrugs.groupby(["Drug"]).sum()["Tumor Volume (mm3)"]
finalVolume
```

```python
# Convert Tumor Size (mm3) into a list for quantitative analysis
tumorSize = filteredLeadingDrugs['Tumor Volume (mm3)'].tolist()
tumorSize
```

```python
# Box plot of the tumor sizes
tumorSizes = filteredLeadingDrugs['Tumor Volume (mm3)']
fig1, ax1 = plt.subplots()
ax1.set_title('Tumor sizes at conclusion of drug regimen')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.boxplot(tumorSizes)
plt.show()

print(f"The list of tumor sizes is: \n \n {tumorSize}")
```

```python
quartiles = tumorSizes.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of tumor volume is: {lowerq}")
print(f"The upper quartile of tumor volume is: {upperq}")
print(f"The interquartile range of tumor volume is: {iqr}")
print(f"The the median of tumor volume is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```
